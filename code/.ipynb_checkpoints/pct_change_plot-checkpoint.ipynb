```python
import pandas as pd
import glob
```

```python
CSV=glob.glob('*csv')
```

```python
CSV
```

```python
S={'pox_year_zlib_sizes.csv':'pox',
 'flu_year_zlib_sizes.csv':'H3N2',
 'hcov_year_zlib_sizes.csv':'hCOV-19',
 'year_zlib_sizes.csv':'RSV'}
```

```python
df=pd.concat([pd.read_csv(x).assign(virus=S[x]) for x in CSV])
```

```python
df["zscore_within_year"] = (
    df.groupby("year")["zlib_bytes"]
      .transform(lambda x: (x - x.mean()) / x.std())
)
```

```python
xf=df[['year','zlib_bytes','virus']]
```

```python
ax=xf[xf.virus=='H3N2'].drop('virus',axis=1).groupby('year').mean().pct_change().plot(style='ko')
xf[xf.virus=='RSV'].drop('virus',axis=1).groupby('year').mean().pct_change().plot(ax=ax,style='ro')
xf[xf.virus=='hCOV-19'].drop('virus',axis=1).groupby('year').mean().pct_change().plot(ax=ax,style='go')
xf[xf.virus=='pox'].drop('virus',axis=1).groupby('year').mean().pct_change().plot(ax=ax,style='cs')
```
